## Speed of the words spoken

**How Speech Speed is Calculated and Graphed from a Transcript:**

To understand how quickly someone is speaking, we analyze the provided transcript, which contains the spoken words and their timings.

1.  **Measuring Speed:** We look at the transcript in short, consecutive time intervals (e.g., every 5 seconds). For each of these intervals, we count the number of words (or characters, for languages like Chinese where words aren't always space-separated). This count is then divided by the duration of the interval (e.g., 5 seconds) to give us a "speed" measurement, typically in words (or characters) per second.

2.  **Plotting the Variation:** The graph visualizes this changing speed. The horizontal axis represents the timeline of the recording. The vertical axis shows the calculated speech speed. Each point on the graph corresponds to the speed measured for a specific time interval. By connecting these points, the line on the graph illustrates how the speaker's pace accelerates or slows down throughout the recording.

This process allows us to see patterns in speaking rhythm and identify periods of faster or slower speech.

```
import matplotlib.pyplot as plt
import json # For the example data

# --- Paste the calculate_words_per_second_windowed function here ---
def calculate_words_per_second_windowed(transcription_json, window_size=5.0, step_size=1.0):
    """
    Calculates words per second (WPS) in a sliding window.

    Args:
        transcription_json (dict): The transcription JSON data.
        window_size (float): The size of the sliding window in seconds.
        step_size (float): The step by which the window slides in seconds.

    Returns:
        list: A list of dictionaries, where each dictionary contains:
              'window_start_time': Start time of the window.
              'window_end_time': End time of the window.
              'words_in_window': Total words from segments starting in this window.
              'wps': Words per second for this window.
    """
    segments = transcription_json.get('segments', [])
    if not segments:
        return []

    results = []
    max_time = 0
    if segments:
        max_time = segments[-1]['end']

    current_window_start = 0.0
    while current_window_start < max_time:
        current_window_end = current_window_start + window_size
        words_in_this_window = 0

        for segment in segments:
            seg_start_time = segment['start']
            if seg_start_time >= current_window_start and seg_start_time < current_window_end:
                text = segment.get('text', '').strip()
                if text:
                    # For Chinese, len(text) counts characters.
                    # If you want space-separated words, use text.split()
                    # For this data, it seems character count is more appropriate as there are few spaces.
                    # Let's assume you want character count for Chinese.
                    # If it's English or space-separated, change to: len(text.split())
                    if any('\u4e00' <= char <= '\u9fff' for char in text): # Check if text contains Chinese characters
                        num_words_in_segment = len(text) # Count characters for Chinese
                    else:
                        num_words_in_segment = len(text.split()) # Count space-separated words otherwise
                    words_in_this_window += num_words_in_segment

        wps = words_in_this_window / window_size if window_size > 0 else 0

        results.append({
            'window_start_time': round(current_window_start, 2),
            'window_end_time': round(current_window_end, 2),
            'words_in_window': words_in_this_window,
            'wps': round(wps, 2) # Words or Characters per second
        })

        current_window_start += step_size

    return results
# --- End of calculate_words_per_second_windowed function ---

# --- You would provide your JSON data below ---
# For example:
transcription_data = result
wps_results = calculate_words_per_second_windowed(transcription_data)
for res in wps_results:
  print(res)
```

```
import matplotlib.pyplot as plt
import json

def plot_wps_variation(wps_results):
    """
    Plots the words per second (WPS) variation over time.

    Args:
        wps_results (list): A list of dictionaries, where each dictionary
                            contains 'window_start_time' and 'wps'.
                            This is the output from calculate_words_per_second_windowed.
    """
    if not wps_results:
        print("No data to plot.")
        return

    # Extract data for plotting
    # We can plot WPS at the start or middle of each window. Let's use start_time.
    window_start_times = [res['window_start_time'] for res in wps_results]
    wps_values = [res['wps'] for res in wps_results]

    # Create the plot
    plt.figure(figsize=(12, 6)) # Adjust figure size as needed
    plt.plot(window_start_times, wps_values, marker='o', linestyle='-', color='b')

    # Add labels and title
    plt.xlabel("Time (seconds) - Start of Window")
    # Update Y-label based on whether you're counting words or characters
    plt.ylabel("Speed (Characters or Words per Second)")
    plt.title("Speech Speed Variation Over Time")
    plt.grid(True) # Add a grid for better readability
    plt.xticks(rotation=45) # Rotate x-axis labels if they overlap
    plt.tight_layout() # Adjust layout to prevent labels from being cut off

    # Show the plot
    plt.show()


# --- Example Usage ---
plot_wps_variation(wps_results)
```

## Pitch variation

## Basic Hardware Requirements for Librosa

- **Processor (CPU):** Any modern CPU capable of running a supported Python version (usually Python 3.6 or later). Librosa itself is a Python package and does not require specialized hardware, but a multi-core processor will improve performance for audio processing tasks.
- **Memory (RAM):** At least 4 GB of RAM is recommended for general audio processing tasks. More RAM may be needed for large audio files or more complex analyses.
- **Storage:** Minimal storage is needed for the library itself (a few megabytes), but sufficient disk space is required to store audio files and any generated data.
- **Operating System:** Librosa supports major OS platforms (Linux, macOS, Windows) where Python and its dependencies can be installed.

**How Audio Pitch is Analyzed and Graphed using Librosa:**

Librosa is a specialized Python library (a toolkit for programmers) designed for analyzing audio and music.

1.  **Understanding Pitch:** "Pitch" is what makes a sound seem high (like a whistle) or low (like a deep hum). In music and speech, it's the fundamental frequency of the sound, often measured in Hertz (Hz).

2.  **Extracting Pitch with Librosa:**
    *   Librosa "listens" to the audio file and breaks it down into many very tiny, consecutive segments of time.
    *   For each of these tiny segments, Librosa uses sophisticated algorithms (like the `piptrack` function in the provided code) to estimate the main, or most prominent, pitch. It's like figuring out the dominant musical note or vocal tone at that exact instant.
    *   If a segment is silent or just noise without a clear tone, Librosa can recognize that no distinct pitch is present.

3.  **Plotting the Pitch Variation:**
    *   The graph visualizes how this detected pitch changes over the entire duration of the audio.
    *   The **horizontal axis** of the graph represents time, from the beginning to the end of the audio.
    *   The **vertical axis** represents the pitch, measured in Hertz (Hz). A higher position on this axis means a higher pitch.
    *   Each point on the plotted line shows the pitch Librosa calculated for a specific moment in the audio. The line connecting these points reveals the melody or intonation contour of the sound—how the pitch rises and falls, creating the tune of a song or the inflection in speech.

In essence, Librosa helps us automatically "transcribe" the pitch information from an audio recording, and the graph provides a clear visual representation of this pitch contour over time.

```
import librosa
# import librosa.display
import matplotlib.pyplot as plt

import numpy as np

# import matplotlib.cm as cm
# cmap = cm.get_cmap('viridis')  # Example usage

# Load the audio file
audio_path = '/content/Easy Cantonese 3 - What do you like about Hong Kong .mp3'  # Replace with your file path
y, sr = librosa.load(audio_path)

# Extract pitch (fundamental frequency) using librosa's piptrack
pitches, magnitudes = librosa.piptrack(y=y, sr=sr)

# Extract the most prominent pitch at each time frame
pitch_values = []
for i in range(pitches.shape[1]):
    index = magnitudes[:, i].argmax()
    pitch = pitches[index, i]
    pitch_values.append(pitch if pitch > 0 else np.nan)  # Ignore 0 Hz (unvoiced)

# Time axis for plotting
frames = range(len(pitch_values))
t = librosa.frames_to_time(frames, sr=sr)

# Plot the pitch variation
plt.figure(figsize=(14, 6))
plt.plot(t, pitch_values, label='Pitch (Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Pitch (Hz)')
plt.title('Pitch Variation Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

## Volume Variation

## Basic Hardware Requirements for Librosa

- **Processor (CPU):** Any modern CPU capable of running a supported Python version (usually Python 3.6 or later). Librosa itself is a Python package and does not require specialized hardware, but a multi-core processor will improve performance for audio processing tasks.
- **Memory (RAM):** At least 4 GB of RAM is recommended for general audio processing tasks. More RAM may be needed for large audio files or more complex analyses.
- **Storage:** Minimal storage is needed for the library itself (a few megabytes), but sufficient disk space is required to store audio files and any generated data.
- **Operating System:** Librosa supports major OS platforms (Linux, macOS, Windows) where Python and its dependencies can be installed.

**How Audio Volume is Analyzed and Graphed using Librosa:**

Librosa is a specialized Python library (a toolkit for programmers) adept at analyzing audio signals.

1.  **Understanding "Volume":** When we talk about the volume of a sound, we're referring to its loudness or intensity. In audio processing, a common way to measure this is by calculating the **Root Mean Square (RMS) energy** of the sound wave.

2.  **Extracting Volume Information with Librosa:**
    *   Librosa first loads the audio file, converting the sound into a digital representation.
    *   It then divides the audio into many very short, consecutive (and often overlapping) segments called "frames."
    *   For each of these tiny frames, Librosa calculates its RMS energy using a function like `librosa.feature.rms`. A frame containing louder sounds will have a higher RMS value, while a quieter frame will have a lower value. This gives a measure of the sound's intensity within that small time window.

3.  **Plotting the Volume Variation:**
    *   The graph visualizes how this calculated RMS energy (our measure of volume) changes over the entire duration of the audio.
    *   The **horizontal axis** of the graph represents time, progressing from the beginning to the end of the audio.
    *   The **vertical axis** represents the volume (RMS energy). A higher point on this axis indicates a louder sound at that moment.
    *   Each point on the plotted line shows the volume calculated for a specific short frame of the audio. The continuous line connecting these points reveals the "volume contour," clearly illustrating how the loudness of the audio rises and falls throughout the recording.

In essence, Librosa processes the audio to extract a detailed, moment-by-moment measure of its volume, and the graph provides an easy-to-understand visual of these loudness fluctuations.

```
import librosa
# import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio file
audio_path = '/content/Easy Cantonese 3 - What do you like about Hong Kong .mp3'  # Replace this with your actual audio file path
y, sr = librosa.load(audio_path)

# Compute RMS (volume) over frames
frame_length = 2048
hop_length = 512
rms = librosa.feature.rms(y=y, frame_length=frame_length, hop_length=hop_length)[0]

# Create time variable for x-axis
frames = range(len(rms))
t = librosa.frames_to_time(frames, sr=sr, hop_length=hop_length)

# Plot the volume variation
plt.figure(figsize=(14, 6))
plt.plot(t, rms, color='purple', label='Volume (RMS Energy)')
plt.xlabel('Time (s)')
plt.ylabel('Volume')
plt.title('Volume Variation Over Time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
